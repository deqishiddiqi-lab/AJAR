#Buatlah sebuah program untuk menambahkan satu atau beberapa mahasiswa dengan
#kriteria sebagai berikut:
#- Program meminta user, berapa data mahasiswa yang akan dimasukkan
#- Program meminta data mahasiswa yang akan disimpan ke dalam dictionary
#dengan struktur:
#- Key : menggunakan NIM singkat (Contoh: 24-009)
#- Value : dengan format
#- Nama (string)
#- Prodi (string)
#- Angkatan (integer)
#- IPK (float)
#- Tambahkan data-data yang telah dimasukkan oleh user ke dalam dictionary.
#- Tampilkan mahasiswa yang telah ditambahkan kedalam dictionary dengan
# menggunakan looping.



```python
angka = int(input("berapa mahasiswa ingin ditambahkan?: "))
output = 0
c = 0
mahasiswa_list = []
for output in range(0, angka,1):

    a = input("masukkan NIM: ")
    b = input("masukkan Nama: ")
    c = input("masukkan prodi: ")
    d = input("masukkan angkatan: ")
    e = float(input("masukkan IPK: "))
    print()
    mahasiswa={
        'NIM':a,
        'Nama':b,
        'prodi':c,
        'angkatan':d,
        'IPK':e
    }
    mahasiswa_list.append(mahasiswa)
print('============================')
for i in mahasiswa_list:
    print()
    print(F'NIM : {i['NIM']}')
    print(F'Nama : {i['Nama']}')
    print(F'prodi : {i['prodi']}')
    print(F'angkatan : {i['angkatan']}')
    print(F'IPK : {i['IPK']}')
print('============================')
```

#Seorang dosen ingin menganalisis nilai praktikum mahasiswa. Buatlah program yang mengelola dan memfilter nilai-nilai ini. Nb: Dilarang menggunakan fungsi/sintaks selain in, append(), len(), dan insert() untuk memanipulasi list.
#- Buatlah sebuah list bernama daftar_nilai yang berisi sepuluh (10) angka bulat acak, yaitu [80, 90, 65, 34, 40, 50, 70, 85, 44, 20].
#- Buatlah dua buah list baru bernama nilai_lulus dan nilai_gagal.
#- Gunakan konsep perulangan untuk mengiterasi seluruh elemen di daftar_nilai.
#- Jika nilai >= 70, masukkan nilai tersebut ke dalam nilai_lulus.
#- Jika nilai < 70, masukkan nilai tersebut ke dalam nilai_gagal.
#- Terdapat beberapa output yang ditampilkan, yaitu:
#- List dari daftar_nilai, nilai_gagal, dan nilai_lulus
#- Nilai rata-rata dari masing-masing daftar_nilai, nilai_gagal, dan nilai_lulus.
#- Total jumlah mahasiswa yang lulus & gagal.(Hint: Gunakan method len() untuk menghitung panjang dari tipe data koleksi)

```python
nilai = [80,90,65,34,40,50,70,85,44,20]
lulus=[]
tidak_lulus=[]
a = 0
b = 0
c=0
for daftar in nilai:
    a +=daftar
for n in nilai:
    if n%2==0:
      tidak_lulus.append(n)
      c +=n
    else:
      lulus.append(n)
      b +=n

print(f'daftar nilai: {nilai}')
print(f'nilai lulus: {lulus}')
print(f'nilai gagal: {tidak_lulus}')
print()
print('nilai rata-rata daftar nilai:',a/len(nilai))
print('nilai rata-rata lulus:',b/len(lulus))
print('nilai rata-rata gagal:',c/len(tidak_lulus))
print()
print("jumlah mahasiswa yang lulus:",len(lulus))
print("jumlah mahasiswa yang gagal:",len(tidak_lulus))
```
